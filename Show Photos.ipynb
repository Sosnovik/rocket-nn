```python
from PIL import Image
import numpy as np
import glob
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w_files = glob.glob('Seq/Seq1M/Seq_*w.pgm')
files = glob.glob('Seq/Seq1M/Seq_*.pgm')
files = [f for f in files if not f in w_files]
```

```python
images = (Image.open(path) for path in files)
im_arrays = np.array([np.array(im) for im in images])
im_arrays = im_arrays.astype(np.float)
im_arrays /= im_arrays.max()
```

```python
def show(image):
    plt.imshow(image, cmap='gray', interpolation='nearest')
    plt.show()
```

```python
bg = im_arrays.mean(axis=0)
show(bg)
```

## Mean

```python
gamma = im_arrays[0].mean()
gamma /= bg.mean()
show(im_arrays[0] - bg * gamma)
```

```python
plt.hist(bg[::5, ::5].ravel(), bins=30)
plt.grid()
plt.show()
```
