```python
def gauss_image(mu_x, mu_y, sigma, size):
    X, Y = np.indices(size)
    # compute gauss function
    arg = (X - mu_x)**2 + (Y - mu_y)**2
    arg = arg.astype('float')
    arg /= 2.0
    arg /= sigma**2
    exp = np.exp(- arg)
    norm = 1 / np.sqrt(2 * np.pi )
    norm /= sigma
    return norm * exp

def random_coordinates(size):
    '''
    Returns random coordinates.....
    '''
    coordinates = []
    for mode in size:
        point = np.random.randint(0, mode)
        coordinates.append(point)
    return coordinates
```

```python
def scaled_random_gauss(sigma, size, scale_factor):
    h_new = size[0]*scale_factor
    w_new = size[1]*scale_factor
    upscaled_size = (h_new, w_new)
    mu_x, mu_y = random_coordinates(upscaled_size)
    upscaled_gauss = gauss_image(mu_x, mu_y, sigma * scale_factor, upscaled_size)
    gauss_img = Image.fromarray(upscaled_gauss)
    original_img = gauss_img.resize(size, Image.ANTIALIAS) 
    arr = np.array(original_img)
    return arr
```

```python
size = (10, 10)
scale_factor = 5
sigma = 1
h_new = size[0]*scale_factor
w_new = size[1]*scale_factor
upscaled_size = (h_new, w_new)
mu_x, mu_y = random_coordinates(upscaled_size)
upscaled_gauss = gauss_image(mu_x, mu_y, sigma * scale_factor, upscaled_size)
gauss_img = Image.fromarray(upscaled_gauss)
original_img = gauss_img.resize(size, Image.ANTIALIAS) 
```

```python
def sample_image(original_image, p_rocket, sigma_rocket, amp_rocket, amp_noise, scale_factor):
    size = original_image.shape[:2]
    prob = np.random.uniform()
    new_image = original_image.copy()
    if prob < p_rocket:
        rocket = scaled_random_gauss(sigma_rocket, size, scale_factor)
        new_image += amp_rocket * rocket
    noise = np.random.normal(size=size)
    new_image += amp_noise * noise
    return new_image 
```
